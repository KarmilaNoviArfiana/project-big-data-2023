```python
# Exchange Rate to USD
sns.set_style('whitegrid')
plt.figure(figsize = (8,5))
sns.lineplot(x = 'year', y = 'exch_usd', hue = 'country', data = df, palette = 'bright')
plt.xlabel('Year')
plt.ylabel('Exchange Rate')
display()
```

```python
#Visualizing different types of crisis
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(18,12), dpi= 60)
axes = axes.flatten()
cols = ['currency_crises','inflation_crises','banking_crisis','systemic_crisis']
for i, ax in zip(cols, axes):
  sns.countplot(y = 'country', ax = ax, data = df, hue = i, palette = 'bright')
plt.tight_layout()
display()
```
