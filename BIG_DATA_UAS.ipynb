```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum as Fsum

# Create a SparkSession
spark = SparkSession.builder.appName("Data Analysis").getOrCreate()

# Read the data into a DataFrame
df = spark.read.format("csv").option("header", True).load("/content/drive/MyDrive/D4 - TI ( SMT 1-8)/SMT 6/BIG DATA/african_crises.csv")

# Calculate the number of null values in each column
null_counts = df.select([Fsum(col(c).isNull().cast("int")).alias(c) for c in df.columns])

# Show the null counts
null_counts.show()

# Stop the SparkSession
#spark.stop()
```
